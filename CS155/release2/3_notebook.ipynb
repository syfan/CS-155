```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
alphas = np.linspace(0.01, 3, 30)
points = []
for val in alphas:
    lasso = Lasso(alpha=val)
    lasso.fit(X_train, y_train)
    points.append(lasso.coef_)
points = np.asarray(points)
plt.figure()
for i in range(9):
    plt.plot(alphas, points[:,i], label=str(i+1))
plt.legend(loc = 1)
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Lasso weights vs alpha')
```

```python
alphas = np.linspace(1, 1e4, 1e4)
points = []
for val in alphas:
    ridge = Ridge(alpha=val)
    ridge.fit(X_train, y_train)
    points.append(ridge.coef_)
points = np.asarray(points)
plt.figure()
for i in range(9):
    plt.plot(alphas, points[:,i], label=str(i+1))
plt.legend(loc = 1)
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge weights vs alpha')
```
